```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Discharge

```python
discharge_notes = pd.read_csv("../../../../../mimic-iv-note-deidentified-free-text-clinical-notes-2.2/note/discharge.csv")
discharge_notes.head()
```

```python
print(discharge_notes['text'][0])
```

```python
length = 0
for i in range(len(discharge_notes)):
    length += len(discharge_notes['text'][i])
print(length/len(discharge_notes))
```

```python
subjectids = len(set(discharge_notes['subject_id']))
hadmids = len(set(discharge_notes['hadm_id']))
noteids = len(set(discharge_notes['note_id']))

print(subjectids, hadmids, noteids)
```

```python
print(hadmids/subjectids)
print(noteids/subjectids)
```
